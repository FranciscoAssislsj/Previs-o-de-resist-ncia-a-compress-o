# Tratamento de Dados

```python
df_concreto = df_concreto.dropna()
```

```python
df_concreto.isnull().sum()
```

```python
df_concreto = df_concreto.drop_duplicates()
```

```python
df_concreto.duplicated().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_concreto.drop(["Concrete compressive strength", "Strength Category"], axis = 1 )

y = df_concreto["Concrete compressive strength"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 25)
```

# Construção e Avaliação de Modelos de Regressão

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, root_mean_squared_error, mean_squared_error, mean_absolute_percentage_error
```

```python
def regression_report(y_true, y_pred):
  mae = mean_absolute_error(y_true, y_pred)
  mse = mean_absolute_error(y_true, y_pred)
  rmse = root_mean_squared_error(y_true, y_pred)
  mape = mean_absolute_percentage_error(y_true, y_pred)
  r2 = r2_score(y_true, y_pred)

  return pd.Series({
      "MAE":mae,
      "MSE":mse,
      "RMSE":rmse,
      "MAPE": mape*100,
      "R2": r2
  })
```

```python
def plot_residual(y_true, y_pred):
  residuals = y_true - y_pred

  plt.scatter(y_pred, residuals)
  plt.axhline(y=0, color="r", linestyle = "--")

  plt.xlabel("Predictions")
  plt.ylabel("Residuals")
  plt.show()
```

```python
model = DecisionTreeRegressor(random_state = 25)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

regression_report(y_test, y_pred)
```

```python
plot_residual(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=5)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

regression_report(y_test, y_pred)
```

```python
plot_residual(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state = 25)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

regression_report(y_test, y_pred)
```

```python
plot_residual(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor(random_state = 25)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

regression_report(y_test, y_pred)
```

```python
plot_residual(y_test, y_pred)
```

```python
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(random_state = 25)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

regression_report(y_test, y_pred)
```

```python
plot_residual(y_test, y_pred)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

regression_report(y_test, y_pred)
```

## Comparando Modelos

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "k-NN": KNeighborsRegressor(),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "MLP (Neural Network)": MLPRegressor(random_state=42)
}
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
results = {}

for model_name, model in models.items():
    print(model_name)

    # MAE
    mae_scores = cross_val_score(model, X_train, y_train, cv=5, scoring="neg_mean_absolute_error")
    mean_mae = np.mean(-mae_scores)

    # R²
    r2_scores = cross_val_score(model, X_train, y_train, cv=5, scoring="r2")
    mean_r2 = np.mean(r2_scores)

    results[model_name] = {
        "Mean MAE": mean_mae,
        "Mean R2": mean_r2
    }

df_results = pd.DataFrame(results).T
print(df_results)
```

```python
df_results.sort_values(by= "Mean R2",ascending = False)
```

```python
df_plot = df_results.sort_values(by= "Mean R2",ascending = False)
```

```python
plt.figure(figsize = (6,6))
plt.barh(df_plot.index, df_plot["Mean MAE"])
plt.xlabel("Mean Absolute Error")
plt.show()
```
